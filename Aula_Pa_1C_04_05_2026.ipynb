```python
# Escrever um programa que o usuário digita um número e o programa imprimi de 0 até o número

A= int(input("Digite o seu número:  "))

for i in range(A + 1):
    print(i)
```

```python
# Escrever um programa que o usúario digita um número e o programa imprime em ordem decrescente
# até o zero

A = int(input("Digite o seu número:  "))

for i in range(A, -1, -1):
    print(i)
```

```python
# Peça um número e mostre todos os números pares até o número

entrada = input("Digite um número inteiro positivo: ")


try:
    numero_limite = int(entrada)

    print(f"Números pares de 0 até {numero_limite}:")

    for i in range(0, numero_limite + 1, 2):
        print(i)

except ValueError:
    print("Por favor, digite um número inteiro válido.")
```

```python
# Peça um número e calcule a soma de 1 até o número


numero = int(input("Digite um número inteiro positivo: "))

soma = (numero * (numero + 1)) // 2

print(f"A soma de 1 até {numero} é: {soma}")
```

```python
# Peça uma palavra e conte as letras: Ex: pyton -A palavra tem 06 letras

palavra = input("Digite uma palavra: ")


quantidade = len(palavra)


print(f"A palavra '{palavra}' tem {quantidade:02d} letras.")

```

```python
# Peça uma palavra e conte as vogais

palavra = input("Digite uma palavra: ")


vogais = "aeiouAEIOU"


total_vogais = sum(1 for letra in palavra if letra in vogais)


print(f"A palavra '{palavra}' tem {total_vogais} vogais.")
```

```python
# Peça ao usuário para digitar 05 números e informe o maior

numeros = []


for i in range(5):

    num = float(input(f"Digite o {i+1}º número: "))
    numeros.append(num)


maior_numero = max(numeros)


print(f"O maior número informado foi: {maior_numero}")
```
