```python
#Escrever um programa que verifica se o usuario é maior de idade

nome = str(input("digite seu nome:  "))
idade = int(input("Digite sua idade:  "))

if idade >= 18:
    print(f"{nome}, é maior de idade")

else:
    print(f"{nome}, não é maior de idade")
```

```python
#Faça um programa que verifica se um número é par ou impar

Número = int(input("Digite seu numero:  "))

if Número % 2 == 0:
    print(f"{Número} é par")

else:
    print(f"{Número} é impar")
```
